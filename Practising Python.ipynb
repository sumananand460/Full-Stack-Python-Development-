# <b style="color: Red">Loops: </b>
<b><p>Loops let you repeat the code writing over again and again (multiple times until you set some limit)</p></b>
<b><p>Loops can be divided into 2 types: 
- 'for' loop and 'while' loop.</p></b>

## <b style="color: Red">For Loop: </b>
<b><p>For loops can be used when you know how many times you have to repeat the code.</p>

## <b style="color: Red">While Loop: </b>
<b><p>While loops can be used when you want to repeat until something changes.</p>

```python
# For loop
fruits = ["apple", "orange", "pineapple", "mango"]

for fruit in fruits:
    print(fruit)
```

```python
# Example:
for i in range(5):
    print(i)
```

```python
# Example 2:
for i in range(0,10,2):
    print(i) # It will print the even number from 0 to 10
```

```python
# Example 3:
word = "SUMAN"
for letters in word:
    print(letters)
```

```python
# Using loops to print numbers from 1 to 10
for i in range (1,11):
    print (i)
```

```python
# Using loop to print hello for 5 times
for i in range(5):
    print("Hello")
```

```python
# Using loop to find the sum of first ten natural numbers
total = 0
for i in range(1,11):
    total = total + i
print(total)
```

```python
# WAP to print numbers from 1 to 50
for i in range (1,51):
    print(i)
```

```python
# WAP to print numbers from 50 to 1 (reverse)
for i in range(50,0,-1):
    print(i, end = " ") # Using 'end' to arrange the output in a single line
```

```python
# WAP to print "Good Morning" for 10 times
for i in range(10):
    print("Good Morning")
```
